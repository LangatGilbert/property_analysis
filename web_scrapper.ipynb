```python
import requests
from requests import get
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = "https://www.the-star.co.ke/classifieds/house-apartment-for-sale/nairobi?s=price-desc&p=2&ei=MtoqW6IMvUO2DmztXWxy4w"

results = requests.get(url)
```

```python
soup = BeautifulSoup(results.text, "html.parser")

#print(soup.prettify())
```

```python
#initialize empty lists where you'll store your data
title = []
location = []
price = []
price_per_sqft = []
agent = []
web_link = []
size =[]
bathrooms = []
description = []
```

```python
property_div = soup.find_all('div', class_='product-info')
```

```python
for container in property_div:
    name = container.find('h2', attrs={"class": "product-title clearfix"}).text
    title.append(name)
    
    locations = container.find('p', attrs={"class": "product-top-meta"}).text
    location.append(locations)
    
    bedrooms = container.find('div', attrs={"class": "product-right-meta"}).text if container.find('div', class_='product-right-meta') else '-'
    size.append(bedrooms)
    
    desc = container.find('div', attrs={"class": "product-rest-info"}).text if container.find('div', class_='product-rest-info') else '-'
    description.append(desc)
    
    posted_date = container.find('span', attrs={"class": "product-date"}).text
    web_link.append(posted_date)
    
    source = container.find('span', attrs={"class": "product-source"})
    agent.append(source)
    
    cost = container.find('div', attrs={"class": "product-price"}).text
    price.append(cost)
    
```

```python
properties = pd.DataFrame({
'title': title,
'location': location,
'size': size,
'description': description,
'posted_date': web_link,
'agent': agent,
'price': price,
})
```

```python
#initialize empty lists where you'll store your data
titles = []
locate = []
prices = []
area = []
agents = []
web_links = []
sizes =[]
bathroom = []
bedroom = []
descriptions = []
```

```python
%%time
n_pages = 0
for page in range(1,400): #the no. of pages to be extracted considering the pages available in the url
    n_pages +=1
    url = 'https://www.buyrentkenya.com/property-for-sale/nairobi?'+'&page='+str(page)
    buyrent_results = requests.get(buyrent_url)
    
    soup = BeautifulSoup(buyrent_results.text, "html.parser")
    
    property_divs = soup.find_all('div', class_='mb-3 w-full')
    
    for container in property_divs:
        names = container.h3.a.text
        titles.append(names)

        locations = container.find('div', attrs={"class": "w-auto flex items-center mb-3 md:mb-2 md:mt-1"}).text if container.find('div', class_='w-auto') else '-'
        locate.append(locations)

        bedrooms = container.find('span', attrs={"class": "text-sm mr-5","data-cy":"card-beds"}).text if container.find('span', attrs={"class": "text-sm mr-5","data-cy":"card-beds"}) else '-'
        bedroom.append(bedrooms)

        bathrooms = container.find('div', attrs={"class": "flex items-center","data-cy":"card-baths"}).text if container.find('div', attrs={"class": "flex items-center","data-cy":"card-baths"}) else '-'
        bathroom.append(bathrooms)

        area_per = container.find('span', attrs={"class": "text-sm","data-cy":"card-area"}).text if container.find('span', attrs={"class": "text-sm","data-cy":"card-area"}) else '-'
        area.append(area_per)

        desc = container.h5.a.text if container.h5.a else '-'
        descriptions.append(desc)

    #     select_one('td.ac > img')['alt']
        source = container.select_one('a.no-underline > img')['alt'] if container.select_one('a.no-underline > img') else '-'
        agents.append(source)

        cost = container.p.a.text
        prices.append(cost)
    
    
    
    
```

```python
buyrent_properties = pd.DataFrame({
'property_title': titles,
'property_location': locate,
'property_area': area,
'property_bedrooms': bedroom,
'property_bathrooms': bathroom,
'property_description': descriptions,
#'posted_date': web_link,
'property_agent': agents,
'property_price': prices,
})
```
